```python
import csv
from collections import defaultdict

def read_data_from_svc(file_name):
    data = []
    with open(file_name, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter='|')
        for row in reader:
            order_number, order_date, product_name, category, quantity, price, total_price = [item.strip() for item in row]
            if (order_number.isnumeric() and not(product_name.isnumeric()) and not(category.isnumeric()) and quantity.isnumeric() and price.isnumeric() and total_price.isnumeric()):
                data.append({'order_number': order_number,'order_date': order_date,'product_name': product_name,'category': category,'quantity': int(quantity),'price': int(price),'total_price': int(total_price)
                })
            else:
                print("Ошибка ввода данных")

    return data

def find_profit(data):
    total_revenue=0
    for datarow in data:
        total_revenue+=datarow["total_price"]
    return total_revenue

def find_most_sold_product(data):
    products_sold = defaultdict(int)
    for item in data:
        products_sold[item['product_name']] += item['quantity']
    n = len(products_sold)
    sold_data =  [0] * n
    print(sold_data)
    i=0
    for item in products_sold.values():
        sold_data[i]=item
        i+=1
    most_sold_prod =''
    max_quant=0
    for i in range(n-1):
        for j in range(n-i-1):
            if sold_data[j] > sold_data[j+1]:
                sold_data[j], sold_data[j+1]= sold_data[j+1], sold_data[j]
    max_quant=sold_data[n-1]
    for key, item in products_sold.items():
        if item==max_quant:
            most_sold_prod+=key
        elif item==max_quant and most_sold_prod!="":
            most_sold_prod+=', '+key
    return most_sold_prod

def find_revenue_product( data):
    n = len(data)
    reven_data = [[0] * 2 for _ in range(n)]
    i = 0
    for key, items in data.items():
        reven_data[i][0] = key
        reven_data[i][1] = items
        i += 1
    return reven_data


def print_report(data):
    profit = find_profit(data)
    print("Общая выручка магазина:", profit)

    most_sold_product = find_most_sold_product(data)
    print("Самый продаваемый товар:", most_sold_product)

    products_revenue= defaultdict(int)
    for item in data:
        products_revenue[item['product_name']] += item['total_price']
    reven_data=find_revenue_product(products_revenue)
    most_reven_prod = ''
    max_quant = 0
    n=len(products_revenue)
    for i in range(n):
        if reven_data[i][1] > max_quant:
            most_reven_prod = reven_data[i][0]
            max_quant = reven_data[i][1]
    print("Товар, принесший наибольшую выручку:", most_reven_prod)
    products_quant = defaultdict(int)

    print("Отчет:")
    for datarow in data:
        products_quant[datarow['product_name']] += datarow['quantity']
    for key, item in products_quant.items():
        revenue=0
        for datarow in data:
            if datarow['product_name']==key:
                revenue+=datarow['total_price']
        profit_rate=(revenue/profit)
        print("|| Наименование товара: " + str(key) + " || Продано единиц товара: " + str(item)+" || Выручка:  " + str(revenue)+" || Доля от выручки: " + str(profit_rate) )


file_name = 'sales_data.svc'
data = read_data_from_svc(file_name)
print_report(data)
```
